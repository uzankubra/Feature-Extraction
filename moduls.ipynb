```python
# seaborn kÃ¼tÃ¼phanesi icinden tips verisetini yukleyÄ±nÄ±z 

df= sns.load_dataset("tips")
df
```

```python
# time deÄŸiÅŸkeninin kategorilerine gore (launch, dinner) total_bill degerlerinin toplamÄ±nÄ± min max ve ortalamasÄ±nÄ± bulunuz.

df.groupby("time").agg({"total_bill": ["min", "max", "sum", "mean"]})
```

```python
# launch zamanÄ±na ve kadÄ±n mÃ¼ÅŸterilere ait total bill ve tip deÄŸerlerinin daye gÃ¶re toplamÄ±nÄ±, min, max ve ortalamasÄ±nÄ± bulunuz

df[(df["time"]=="Lunch")& (df["sex"]=="Female")].groupby("day").agg({"total_bill": ["min", "max", "sum", "mean"], "tip": ["min", "max", "sum", "mean"]})
```

```python
# size 3ten kÃ¼Ã§Ã¼k total_bill'i 10dan bÃ¼yÃ¼k olan sipariÅŸlerin ortalamasÄ± nedir

df[(df["size"]<3)&(df["total_bill"]>10)]["total_bill"].mean()
```

```python
# total_bill degiskeninin kadin ve erkek icin ayri ayrÄ± ortalamasÄ±nÄ± bulun

# buldugunuz ortalamanÄ±n altÄ±nda olanlara 0, esit ve ya ustunde olanlara 1 verildigi yeni bir total_bill_flag olusturun

# Dikkat! KadÄ±nlarÄ±n ortalamasÄ± ayrÄ±, erkeklerin ortalamasÄ± ayrÄ± degerlendirilecek

# parametre olarak cinsiyet ve total_bill alan bir fonksiyon yazarak baslayÄ±n 

female_avg= df[df["sex"]=="Female"]["total_bill"].mean()

male_avg= df[df["sex"]== "Male"]["total_bill"].mean()

def function(sex, total_bill):
    if sex== "Female":
        if total_bill >= female_avg:
            return 1
        else:
            return 0
        
    else :
        if total_bill >= male_avg:
            return 1
        
        else:
            return 0
        
        
    
    
```

```python
df["total_bill_flag"]=df.apply(lambda x: function(x["sex"], x["total_bill"]), axis=1)
```

```python
df["total_bill_flag"]
```

```python
#total_bill_flag degiskenini kullanarak cinsiyetlere gÃ¶re ortalamanÄ±n altÄ±nda ve Ã¼stÃ¼nde olanlarÄ±n sayÄ±sÄ±nÄ± gÃ¶zlemleyiniz

df.groupby(["sex", "total_bill_flag"]).agg({"total_bill_flag": "count"})
```

```python
# total_bill_tip_sum degiskenine gore bÃ¼yÃ¼kten kÃ¼Ã§Ã¼ÄŸe sÄ±ralayÄ±nÄ±z ve ilk 30 kisiyi yeni bir dataframe'e atayÄ±nÄ±z

df["total_bill_tip_sum"]= df["total_bill"]+ df["tip"]
```
